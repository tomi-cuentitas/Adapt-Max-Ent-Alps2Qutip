```python
import numpy as np
from scipy.optimize import fsolve

def Heisenberg_vLR(params):
    Jx = params['Jx']
    Jy = params['Jy']
    Jz = params['Jz']
    
    # Correct polynomial: x^3 - (Jx*Jy + Jx*Jz + Jy*Jz) - 2*Jx*Jy*Jz = 0
    poly_coeffs = [1, 0, -(Jx*Jy + Jx*Jz + Jy*Jz), -2*Jx*Jy*Jz]
    roots = np.roots(poly_coeffs)
    Ffactor = np.max(np.real(roots[np.isreal(roots)]))

    # Solve x * arcsinh(x) = sqrt(x^2 + 1)
    chi_y = fsolve(lambda x: x*np.arcsinh(x) - np.sqrt(x**2 + 1), 0.1)[0]
    
    return 4 * Ffactor * chi_y
```

```python
import numpy as np
params={}


params['size']=7
params['Jx']=1.30; params['Jy'] = 0.83*params['Jx']; params['Jz']=0.*params['Jx']

from scipy.optimize import root, fsolve
Ffactor=np.real(max(np.roots(np.poly1d([1, 0, -(params['Jx']*params['Jy']+params['Jx']*params['Jy']+params['Jy']*params['Jz']), 
                           -2*params['Jx']*params['Jy']*params['Jz']]))))
chi_y=fsolve(lambda x,y: x*np.arcsinh(x)-np.sqrt(x**2+1)-y, 1e-1, args=(0))[0]
vLR=4*Ffactor*chi_y
vLR
```

```python
from pathlib import Path
import pandas as pd

def load_and_concat(folder_path: str, label: str = None):
    folder = Path(folder_path)
    csv_files = sorted(folder.glob("traj_*.csv"))

    print(f"\n[INFO] Searching in: {folder}")
    print(f"Found {len(csv_files)} CSV files.")

    if not csv_files:
        print("[WARNING] No files found.")
        return None

    # Load all DataFrames with their row counts
    dfs = []
    for f in csv_files:
        try:
            df = pd.read_csv(f)
            dfs.append((df.shape[0], f.name, df))
            print(f"  Loaded: {f.name} | shape = {df.shape}")
        except Exception as e:
            print(f"[ERROR] Failed to load {f.name}: {e}")

    if not dfs:
        print("[WARNING] No valid DataFrames loaded.")
        return None

    # Select the one with the largest number of rows
    longest_df = max(dfs, key=lambda x: x[0])
    print(f"[SUCCESS] Selected longest file: {longest_df[1]} with {longest_df[0]} rows.")

    return longest_df[2]

# === USER INPUT ===
folder_paths = {
    "ell10_m2_eps0001": "Jx1.3_Jy0.83_Jz0.0_ell10_m2_eps0.001",
    "ell6_m2_eps0001": "Jx1.3_Jy0.83_Jz0.0_ell6_m2_eps0.001",
    "ell6_m3_eps0001": "Jx1.3_Jy0.83_Jz0.0_ell6_m3_eps0.001",
    "ell8_m2_eps0001": "Jx1.3_Jy0.83_Jz0.0_ell8_m2_eps0.001"
}

# === LOAD ALL ===
all_dataframes = {}
for label, path in folder_paths.items():
    df = load_and_concat(path, label)
    if df is not None:
        all_dataframes[label] = df
```

```python
timespan = all_dataframes["ell8_m2_eps0001"]["t"].to_numpy(dtype=float)
ev_ex = all_dataframes["ell8_m2_eps0001"]['ev_exact'].to_numpy(dtype=complex)
ev = all_dataframes["ell8_m2_eps0001"]['ev'].to_numpy(dtype=complex)
```

```python

label_sim1 = fr'Ada-Max-Ent $(\ell, m0, \epsilon) = ({10}, {2}, {0.0001})$'
label_sim2 = fr'Ada-Max-Ent $(\ell, m0, \epsilon) = ({6}, {2}, {0.0001})$'
label_sim3 = fr'Ada-Max-Ent $(\ell, m0, \epsilon) = ({6}, {3}, {0.0001})$'
label_sim4 = fr'Ada-Max-Ent $(\ell, m0, \epsilon) = ({8}, {2}, {0.0001})$'
```

```python
import matplotlib.pyplot as plt
import numpy as np

from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# --- Main figure ---
fig, ax = plt.subplots(figsize=(8, 8))

n = 5
m = 180

# --- Plotting the exact and approximate expectation values ---
ax.plot(timespan[:m][::n] * vLR, ev_ex[:m][::n],
        label='Exact', color="blue", linestyle='dashdot', linewidth=1)

ax.plot(timespan[:m][::n] * vLR, ev_maxent_ell10_m2_eps0001[:m][::n],
        label=label_sim1, color='red', linestyle='--', marker='D', linewidth=1)

ax.plot(timespan[:m][::n] * vLR, ev_maxent_ell6_m2_eps0001[:m][::n],
        label=label_sim4, color='orange', linestyle='--', marker='D', linewidth=1)


ax.plot(timespan[:m][::n] * vLR, ev_maxent_ell8_m2_eps0001[:m][::n],
        label=label_sim2, color='green', linestyle='--', marker='D', linewidth=1)

ax.plot(timespan[:m][::n] * vLR, ev_maxent_ell6_m3_eps0001[:m][::n],
        label=label_sim3, color='magenta', linestyle='--', marker='D', linewidth=1)

# --- Inset: percentage errors ---
inset_ax = inset_axes(ax, width="40%", height="30%", loc='upper left', borderpad=3)

# Define error function
def percentage_error(exact, approx):
    exact = np.array(exact, dtype=np.complex128)
    approx = np.array(approx, dtype=np.complex128)
    return 100 * np.abs(approx - exact) / np.abs(exact)

# Compute and plot Δ% for each approximation
inset_ax.plot(timespan[:m] * vLR, percentage_error(ev_ex[:m], ev_maxent_ell6_m2_eps0001[:m]),
              color='red', linestyle='-', linewidth=1, label=label_sim1)

inset_ax.plot(timespan[:m] * vLR, percentage_error(ev_ex[:m], ev_maxent_ell8_m2_eps0001[:m]),
              color='green', linestyle='-', linewidth=1, label=label_sim2)

inset_ax.plot(timespan[:m] * vLR, percentage_error(ev_ex[:m], ev_maxent_ell6_m3_eps0001[:m]),
              color='magenta', linestyle='-', linewidth=1, label=label_sim3)

inset_ax.plot(timespan[:m] * vLR, percentage_error(ev_ex[:m], ev_maxent_ell8_m2_eps0001[:m]),
              color='orange', linestyle='-', linewidth=1, label=label_sim4)

inset_ax.set_title(r'$\Delta$ (%)', fontsize=10)
inset_ax.tick_params(labelsize=8)
inset_ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.6)

# --- Style and legend ---
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(direction='out', length=6, width=2, which='major', grid_alpha=0.6)
ax.grid(True, which='major', linestyle='--', linewidth=0.75, alpha=0.7)
ax.legend(frameon=False, loc='center left')

# --- Twin x-axis (top) ---
ax1 = ax.twiny()
ax1.set_xlim(timespan[0], timespan[-1])
ax.set_xlabel(r'$t v_{\rm LR} \, [L]$')
ax1.set_xlabel(r'$t \, [{\rm arb. units}]$')

# --- Title ---
plt.title(r'Non-local Hamiltonian, $L={}$, $J_x={}$, $J_z={}$, $v_{{\rm LR}}={}$'.format(
    7, round(1.5, 3), params["Jz"], round(vLR, 3)))

plt.tight_layout()
plt.show()
```
